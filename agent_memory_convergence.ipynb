It simulates two agents updating their memory from shared observations and visualizes how their memories converge (via cosine similarity) and diverge (via Euclidean distance) over time.

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
from numpy.linalg import norm

# Parameters
timesteps = 100
memory_dim = 50
gamma = 0.1  # plasticity factor

# Initialization
M1 = np.random.rand(memory_dim)
M2 = np.random.rand(memory_dim)

similarities = []
distances = []

for t in range(timesteps):
    # Shared observation with small random noise
    O_t = np.random.rand(memory_dim)

    # Update memories
    M1 = M1 + gamma * (O_t - M1)
    M2 = M2 + gamma * (O_t - M2)

    # Cosine similarity and Euclidean distance
    sim = cosine_similarity([M1], [M2])[0][0]
    dist = norm(M1 - M2)

    similarities.append(sim)
    distances.append(dist)

# Plotting
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].plot(similarities)
ax[0].set_title("Cosine Similarity (Convergence)")
ax[0].set_xlabel("Time")
ax[0].set_ylabel("Similarity")

ax[1].plot(distances)
ax[1].set_title("Euclidean Distance (Divergence)")
ax[1].set_xlabel("Time")
ax[1].set_ylabel("Distance")

plt.tight_layout()
plt.show()
```

Here's a visual plot of two agents updating their internal memories based on shared observations over time. 
- On the left, we see their convergence through increasing cosine similarity. 
- On the right, divergence decreases via normalized Euclidean distance.
